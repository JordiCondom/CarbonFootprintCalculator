```python
############################################### Scraping #############################################################

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium import webdriver

import re
import time

# Define the regular expressions
question_regex = r'<legend class="question-title[^"]*">(.+?)</legend>'
answer_regex = r'<li[^>]*>(.+?)</li>'

# Create the questionnaire dictionary
questionnaire = {}

# Set up Chrome WebDriver
driver = webdriver.Chrome()
url = "https://footprint.wwf.org.uk/questionnaire"

# Navigate to the URL
driver.get(url)

# Wait for the cookie pop-up to appear
time.sleep(2)
# Find and click the "Accept All Cookies" button
accept_button = driver.find_element(By.XPATH, '//button[text()="Accept All Cookies"]')
accept_button.click()

non_finito = True
while non_finito:
#     # Wait for the page to load completely
#     wait = WebDriverWait(driver, 10)
#     wait.until(EC.presence_of_element_located((By.XPATH, '//button[contains(text(), "Next Question")]')))
    # Find all question and answer elements
    question_elements = driver.find_elements(By.XPATH, '//legend[contains(@class, "question-title")]')
    answer_elements = driver.find_elements(By.XPATH, '//ul[contains(@class, "answer-list")]/li[contains(@class, "checkbox-row")]')

    # Extract the question and answer text
    questions = [question_element.text for question_element in question_elements]
    answers = [answer_element.find_element(By.CLASS_NAME, 'answer-text').text for answer_element in answer_elements]
    if len(answers) != 0:
        questionnaire[questions[0]] = answers
    else:
        questionnaire[questions[0]] = 'empty_ans'
    try:
        next_botton = driver.find_element(By.CSS_SELECTOR, 'li[data-index="1"].checkbox-row.mb-6.col-md-12')
        print(f'{q} - ans')
    except:
        try:
            next_botton = driver.find_element(By.XPATH, '//button[contains(text(), "Next Question")]')
            print(f'{q} - nxt')
        except:
            next_botton = driver.find_element(By.XPATH, '//button[contains(text(), "Finish")]')
            print(f'{q} - fin')
            non_finito = False
    next_botton.click()
    driver.forward()
#     driver.refresh()
    # Wait
    time.sleep(3)

for question, answers in questionnaire.items():
    print("Question:", question)
    print("Answers:")

    for answer in answers:
        print("-", answer)
    print()
# Close the browser
driver.quit()
```
